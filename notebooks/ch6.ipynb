## Doc String

```python

def fact(num):
    """ This function calculates factorial.
        It takes any positive number.
        ex = 6 x 5 x 4 x 3 x 2 x 1 = 720
    """
    total = 1
    if num>0:
        while num > 0:
            total = num*total
            num = num-1
        return total
    else:
        return "Please enter positive number only"
    
print(fact(5))
print(fact.__doc__)
```

## Parameter and Arguments

```python
def greet(name): # parameter
    return f"Good Morning {name}"
greet('Himanshu') #argument
```

```python
def greet(name,t): # parameters
    if t>4 and t<12:
        return f"Good Morning {name}"
    elif t>12 and t<16:
        return f"Good Afternoon {name}"
    elif t>16 and t<21:
        return f"Good Evening {name}"
    else:
        return f"Good Night {name}"

#function call
print(greet('Vishal',5)) #arguments
```

## Positional or required arguments


```python
def fun(b,a):
    return a,b
print(fun(4,6)) #positional
print(fun(3)) #required
```

```python
#default argument
def greet(name="User"):
    return f"Hello {name}"
    
print(greet())
```

```python
#keyword arguments
def voteCheck(name, age):
    msg = f"Hello {name}, you are {age} years old\n"
    d1 = "So you can vote"
    d2 = "So you can not vote"
    
    if age> 18:
        return msg + d1
    else:
        return msg + d2
        
res = voteCheck("Geeta",20)
print(res)
```

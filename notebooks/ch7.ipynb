```python
#SHOP
# Predefined fruit price list
fruit_shop = {
    "mango": 30,
    "kiwi": 90,
    "apple": 100,
    "lemon": 30,
    "orange": 50
}
f = open("fruitCart.txt","w")

# Lists to store cart details
cart_items = []
quantities = []

# User input loop
repeat = "y"
while repeat == "y":
    item_name = input("\U0001F34A Enter the fruit you want to buy: ").lower()

    if item_name in fruit_shop:
        cart_items.append(item_name)
        item_quantity = int(input(f"\U0001F4E6 Enter quantity of {item_name}: "))
        quantities.append(item_quantity)
        f.write(f" {item_name.title()} added successfully!\n")   # âœ…
    else:
        f.write(f"Sorry, {item_name} is not available right now.\n")  # âŒ

    repeat = input("\U0001F501 Do you want to buy something else? (y/n): ").lower()

# Display cart summary
f.write("\n Your Cart Summary:")
total_amount = 0
for index in range(len(cart_items)):
    fruit = cart_items[index]
    quantity = quantities[index]
    price = fruit_shop[fruit] * quantity
    total_amount += price
    f.write(f" {fruit.title()} x {quantity} = {price}")

f.write(f"\n Total Amount: {total_amount}")
f.write("\n****** Thanks for shopping with us! ******")

f.close()
```
